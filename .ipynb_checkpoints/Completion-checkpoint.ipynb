```python
#from __future__ import division
import pandas as pd
from datetime import datetime
import numpy as np
import math
from haversine import haversine
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('./Geolife-beijing_original.tsv', header=0,delimiter='\t',encoding='latin1')
```

```python
df.head()
```

```python
df.loc[:, 'date'] = df[['date', 'time']].apply(lambda i: datetime.strptime(i.date+str(' ')+i.time, '%Y-%m-%d %H:%M:%S') , axis=1)
```

Selecting only traces in the city center of Beiging

```python
bottomLeft = (39.441615, 115.417739)
bottomRight = (39.441615, 117.507993)
topLeft = (41.055618, 115.417739)
topRight = (41.055618, 117.507993)
```

```python
######################################### cells 200 m X 200 m #########################################
k = 0.2 #km
cols = np.linspace(bottomLeft[1], bottomRight[1], num = math.ceil(max(haversine(bottomLeft,topLeft),haversine(bottomRight,topRight)) / k))
rows = np.linspace(bottomLeft[0], topLeft[0], num = math.ceil(max(haversine(bottomLeft,bottomRight),haversine(topLeft,topRight)) / k))
df['col'] = np.searchsorted(cols, df['lon'])
df['row'] = np.searchsorted(rows, df['lat'])


######################################### cells 400 m X 400 m #########################################
k = 0.4 #km
cols = np.linspace(bottomLeft[1], bottomRight[1], num = math.ceil(max(haversine(bottomLeft,topLeft),haversine(bottomRight,topRight)) / k))
rows = np.linspace(bottomLeft[0], topLeft[0], num = math.ceil(max(haversine(bottomLeft,bottomRight),haversine(topLeft,topRight)) / k))
df['col_1'] = np.searchsorted(cols, df['lon'])
df['row_1'] = np.searchsorted(rows, df['lat'])


######################################### cells 600 m X 600 m #########################################
k = 0.6 #km
cols = np.linspace(bottomLeft[1], bottomRight[1], num = math.ceil(max(haversine(bottomLeft,topLeft),haversine(bottomRight,topRight)) / k))
rows = np.linspace(bottomLeft[0], topLeft[0], num = math.ceil(max(haversine(bottomLeft,bottomRight),haversine(topLeft,topRight)) / k))
df['col_2'] = np.searchsorted(cols, df['lon'])
df['row_2'] = np.searchsorted(rows, df['lat'])

######################################### cells 800 m X 800 m #########################################
k = 0.8 #km
cols = np.linspace(bottomLeft[1], bottomRight[1], num = math.ceil(max(haversine(bottomLeft,topLeft),haversine(bottomRight,topRight)) / k))
rows = np.linspace(bottomLeft[0], topLeft[0], num = math.ceil(max(haversine(bottomLeft,bottomRight),haversine(topLeft,topRight)) / k))
df['col_3'] = np.searchsorted(cols, df['lon'])
df['row_3'] = np.searchsorted(rows, df['lat'])


######################################### cells 1000 m X 1000 m #########################################
k = 1 #km
cols = np.linspace(bottomLeft[1], bottomRight[1], num = math.ceil(max(haversine(bottomLeft,topLeft),haversine(bottomRight,topRight)) / k))
rows = np.linspace(bottomLeft[0], topLeft[0], num = math.ceil(max(haversine(bottomLeft,bottomRight),haversine(topLeft,topRight)) / k))
df['col_4'] = np.searchsorted(cols, df['lon'])
df['row_4'] = np.searchsorted(rows, df['lat'])


######################################### cells 2000 m X 2000 m #########################################
k = 2 #km
cols = np.linspace(bottomLeft[1], bottomRight[1], num = math.ceil(max(haversine(bottomLeft,topLeft),haversine(bottomRight,topRight)) / k))
rows = np.linspace(bottomLeft[0], topLeft[0], num = math.ceil(max(haversine(bottomLeft,bottomRight),haversine(topLeft,topRight)) / k))
df['col_5'] = np.searchsorted(cols, df['lon'])
df['row_5'] = np.searchsorted(rows, df['lat'])


######################################### cells 4000 m X 4000 m #########################################
k = 4 #km
cols = np.linspace(bottomLeft[1], bottomRight[1], num = math.ceil(max(haversine(bottomLeft,topLeft),haversine(bottomRight,topRight)) / k))
rows = np.linspace(bottomLeft[0], topLeft[0], num = math.ceil(max(haversine(bottomLeft,bottomRight),haversine(topLeft,topRight)) / k))
df['col_6'] = np.searchsorted(cols, df['lon'])
df['row_6'] = np.searchsorted(rows, df['lat'])
```

```python
df.loc[:, 'geohash'] = df[['col','row']].apply(lambda i: str(i.col)+str('+')+str(i.row), axis = 1)
```

```python
df.columns = ['lat','lon','date','dev_id','geohash','geohash_1','geohash_2','geohash_3','geohash_4','geohash_5','geohash_6']
```

```python
fr=df.loc[(df['date'].dt.hour>=2) & (df['date'].dt.hour<=6)] 
a=fr.groupby([fr['dev_id'],fr['geohash']]).count()
a.reset_index(['geohash'],inplace=True)
home=a.groupby('dev_id').max().geohash
```

Identify home location for each user

```python
fr=df.loc[(df['date'].dt.hour>=2) & (df['date'].dt.hour<=6)] 
a=fr.groupby([fr['dev_id'],fr['geohash']]).count()
a.reset_index(['geohash'],inplace=True)
home=a.groupby('dev_id').max().geohash
```

Identify work location for each user (between 10am and 11am)

```python
fr=df.loc[(df['date'].dt.hour>=10) & (df['date'].dt.hour<=11)] 
a=fr.groupby([fr['dev_id'],fr['geohash']]).count()
a.reset_index(['geohash'],inplace=True)
work_1=a.groupby('dev_id').max().geohash
```

Identify work location for each user (between 2pm and 5pm)

```python
fr=df.loc[(df['date'].dt.hour>=14) & (df['date'].dt.hour<=17)] 
a=fr.groupby([fr['dev_id'],fr['geohash']]).count()
a.reset_index(['geohash'],inplace=True)
work_2=a.groupby('dev_id').max().geohash
```

Complete the missing records

```python
df1=df.set_index(['dev_id'])
list_usr=df1.index
list_usr=list_usr.drop_duplicates()
```

```python
tab_frame_home=[]
df_x=df.set_index(['dev_id','geohash'])
for i in list_usr:
    if i in home.keys():
        fr=df1.loc[i]
        fr=fr.loc[(fr['date'].dt.hour>=2) & (fr['date'].dt.hour<=6)] 
        fr=fr.set_index(['date'])
        fr=fr.resample('5T').last()
        fr['geohash'].fillna(home[i], inplace = True) 
        l=df_x.loc[i,home[i]]
        fr['lat'].fillna(l.lat.values[0], inplace = True) 
        fr['lon'].fillna(l.lon.values[0], inplace = True) 
        fr['geohash_1'].fillna(l.geohash_1.values[0], inplace = True) 
        fr['geohash_2'].fillna(l.geohash_2.values[0], inplace = True) 
        fr['geohash_3'].fillna(l.geohash_3.values[0], inplace = True) 
        fr['geohash_4'].fillna(l.geohash_4.values[0], inplace = True) 
        fr['geohash_5'].fillna(l.geohash_5.values[0], inplace = True) 
        fr['geohash_6'].fillna(l.geohash_6.values[0], inplace = True) 
        
        fr.loc[:, 'dev_id'] =fr.apply(lambda x: i, axis=1)
        if len(tab_frame_home)==0:
            tab_frame_home=fr
        else:
            tab_frame_home=pd.concat([tab_frame_home, fr], axis=0, sort=False)
```

```python
len(tab_frame_home)
```

```python
df1=df.set_index(['dev_id'])

for i in list_usr:
    if i in work_1.keys():
        fr=df1.loc[[i]]
        fr=fr.loc[(fr['date'].dt.hour>=10) & (fr['date'].dt.hour<=11)] 
        fr=fr.set_index(['date'])
        fr=fr.resample('5T').last()
        fr['geohash'].fillna(work_1[i], inplace = True) 
        l=df_x.loc[i,work_1[i]]
        fr['lat'].fillna(l.lat.values[0], inplace = True)
        fr['lon'].fillna(l.lon.values[0], inplace = True)
        fr['geohash_1'].fillna(l.geohash_1.values[0], inplace = True)
        fr['geohash_2'].fillna(l.geohash_2.values[0], inplace = True)
        fr['geohash_3'].fillna(l.geohash_3.values[0], inplace = True)
        fr['geohash_4'].fillna(l.geohash_4.values[0], inplace = True)
        fr['geohash_5'].fillna(l.geohash_5.values[0], inplace = True)
        fr['geohash_6'].fillna(l.geohash_6.values[0], inplace = True)
        fr.loc[:, 'dev_id'] =fr.apply(lambda x: i, axis=1)
        tab_frame_home=pd.concat([tab_frame_home, fr], axis=0, sort=False)
```

```python
df1=df.set_index(['dev_id'])

for i in list_usr:
    if i in work_2.keys():
        fr=df1.loc[[i]]
        fr=fr.loc[(fr['date'].dt.hour>=14) & (fr['date'].dt.hour<=17)] 
        fr=fr.set_index(['date'])
        fr=fr.resample('5T').last()
        fr['geohash'].fillna(work_2[i], inplace = True) 
        l=df_x.loc[i,work_2[i]]
        fr['lat'].fillna(l.lat.values[0], inplace = True) 
        fr['lon'].fillna(l.lon.values[0], inplace = True) 
        fr['geohash_1'].fillna(l.geohash_1.values[0], inplace = True) 
        fr['geohash_2'].fillna(l.geohash_2.values[0], inplace = True) 
        fr['geohash_3'].fillna(l.geohash_3.values[0], inplace = True) 
        fr['geohash_4'].fillna(l.geohash_4.values[0], inplace = True) 
        fr['geohash_5'].fillna(l.geohash_5.values[0], inplace = True) 
        fr['geohash_6'].fillna(l.geohash_6.values[0], inplace = True) 
        
        fr.loc[:, 'dev_id'] =fr.apply(lambda x: i, axis=1)
        tab_frame_home=pd.concat([tab_frame_home, fr], axis=0, sort=False)
```
